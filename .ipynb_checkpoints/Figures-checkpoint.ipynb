```python
env =  "gibson" #"doom" # 
p_values_nav = pd.read_csv("results/{}_navigate_reversal_sig.csv".format(env))
nav_color = (142, 210, 254, 1)
p_values_explore = pd.read_csv("results/{}_explore_reversal_sig.csv".format(env))
explore_color = (246, 146, 143, 1)
```

```python
def get_std_name(task):
    if task == 'class_1000':
        return 'class1000'
    if task == 'class_places':
        return 'classplaces'
    return task
```

```python
# Create the graph from the above files
p_value_to_use = 'p' #'adjusted_p' # 
G = nx.MultiDiGraph()
G.add_nodes_from(['segmentsemantic', 'class1000', 'classplaces', 'rgb2mist', 'rgb2mist'])
for i, r in p_values_nav.iterrows():
    edge = tuple(r['feature'].split("_h1_vs_h0_")[::-1])
    edge = tuple([get_std_name(e) for e in edge])
    G.add_edge(*edge, attr={'weight': r[p_value_to_use], 'color': nav_color})

# for i, r in p_values_explore.iterrows():
#     edge = tuple(r['feature'].split("_h1_vs_h0_")[::-1])
#     edge = tuple([get_std_name(e) for e in edge])
#     G.add_edge(*edge, attr={'weight': r[p_value_to_use], 'color': explore_color})
```

```python
# Plot the graph
def get_lw(p):
    if 0 < p < 0.001:
        return 6
    elif 0.001 < p < 0.01:
        return 4
    elif 0.01 < p < 0.05:
        return 2
    else:
        return 0

def get_horizontal_alignment(node):
    mapping = {'class1000': 'right',
               'classplaces': 'center',
               'segmentsemantic':'right',
               'reshade':'center',
               'keypoint3d':'left',
               'rgb2mist':'left',}
    return mapping[node]

def get_vertical_alignment(node):
    mapping = {'class1000': 'bottom',
               'classplaces': 'top',
               'segmentsemantic':'top',
               'reshade':'bottom',
               'keypoint3d':'bottom',
               'rgb2mist':'top',}
    return mapping[node]

def convert_to_hex(rgba_color) :
    red = int(rgba_color[0])
    green = int(rgba_color[1])
    blue = int(rgba_color[2])
    return '#%02x%02x%02x' % (red, green, blue)

def draw_network(G,pos,ax,sg=None):

    for n in G:
        print(pos[n])
#             '3d': "#FB6F63",
#     'semantic': '#2B8FE3',
        print(n)
        c=Circle(pos[n],radius=0.07,alpha=1.0, color='#FB6F63' if n in ['keypoint3d','rgb2mist','reshade'] else '#2B8FE3' )
        ax.add_patch(c)
        G.node[n]['patch']=c
        vadjustment = 0.1 if get_vertical_alignment(n) == 'bottom' else -0.1
        hadjustment = 1.2 if abs(pos[n][1]) < 1e-5 else 1.0
        ax.text(pos[n][0] * hadjustment, pos[n][1] + vadjustment * hadjustment, map_to_display_names[n],
#                 weight='light', alpha=1.0, color='k',
                horizontalalignment=get_horizontal_alignment(n),
                verticalalignment=get_vertical_alignment(n))
    seen={}
    for (u,v,d) in G.edges(data=True):
        n1=G.node[u]['patch']
        n2=G.node[v]['patch']
        color = d['attr']['color']
        rad=0.1
        if (u,v,color) in seen:
            rad=seen.get((u,v,color))
            rad=(rad+np.sign(rad)*0.1)*-1
        alpha=1.0
        color='k'

        e = FancyArrowPatch(n1.center,n2.center,patchA=n1,patchB=n2,
                            arrowstyle='-|>',
#                             connectionstyle='arc3,rad=%s'%rad,
                            mutation_scale=30.0,
                            lw=get_lw(d['attr']['weight']),
                            alpha=alpha,
                            color=convert_to_hex(d['attr']['color']))
        seen[(u,v)]=rad
        if get_lw(d['attr']['weight']) > 0:
            ax.add_patch(e)
    return e


# nodes = ('class1000', 'segmentsemantic', 'reshade', 'classplaces',  'rgb2mist', 'keypoint3d')
nodes = ('reshade', 'keypoint3d', 'class1000', 'classplaces', 'segmentsemantic',   'rgb2mist',)
pos=nx.circular_layout(enumerate(nodes))
pos = {nodes[i]: v for i, (k, v) in enumerate(pos.items())}
plt.figure(figsize=(4,4))
ax=plt.gca()
draw_network(G,pos,ax)
ax.autoscale()
plt.axis('equal')
plt.axis('off')
plt.savefig('figures/{}_rankreversal.pdf'.format(env), transparent=True)
plt.show()
```

```python
nav_settings = {'gibson_file': "data/gibson/gibson_navigate.csv",
                'gibson_timestep': 420,
                'doom_file': 'data/doom/doom_nav.csv',
                'doom_timestep': 840,
                'title': "Feature Correlation (Navigate)",
                'out_file': 'figures/navigate_doomvgibson.pdf'
               }

explore_settings = {'gibson_file': "data/gibson/gibson_explore.csv",
                'gibson_timestep': 480,
                'doom_file': 'data/doom/doom_explore.csv',
                'doom_timestep': 1020,
                'title': "Feature Correlation (Explore)",
                'out_file': 'figures/explore_doomvgibson.pdf'
               }
nav_sig = {'class1000',  'segmentsemantic', 'curvature', 'classplaces', 'segment25d',  'segment2d', 'rgb2mist', 'edge3d',
           'vanishingpoint', 'reshade', 'rgb2sfnorm', 'roomlayout' , 'autoencoder', 'jigsaw'}
exp_sig = set(['reshade', 'keypoint3d', 'curvature', 'rgb2mist', ])
```

```python
# settings = nav_settings
# settings = explore_settings

for plot_num, settings in enumerate([nav_settings, explore_settings]):
    df = pd.read_csv(settings['gibson_file'])
    df = df.replace('vanishingpoint', 'vanishing_point')
    df = df.replace('inpaintingwhole', 'inpainting_whole')
    df = df.replace('roomlayout', 'room_layout')
    df = df.replace('random_projection', 'random')

    selected_ts = df[(df['timestep'] == settings['gibson_timestep']) & (df['train'] == False) ].groupby(['feature', 'seed']).mean()
    gibson_df = selected_ts.groupby(['feature']).mean().sort_values('reward', ascending=False)

    df = pd.read_csv(settings['doom_file'])
    df = df.replace('class_1000', 'class1000')
    df = df.replace('class_places', 'classplaces')
    selected_ts = df[(df['timestep'] == settings['doom_timestep']) & (df['train'] == False) ].groupby(['feature', 'seed']).mean()
    doom_df = selected_ts.groupby(['feature']).mean().sort_values('reward', ascending=False)

    results = gibson_df.merge(doom_df, on=['feature'], suffixes=('_gibson', '_doom'))
    results = results.sort_values('reward_doom', ascending=False)
    results['rank_doom'] = range(len(results))
    results = results.sort_values('reward_gibson', ascending=False)
    results['rank_gibson'] = range(len(results))

    plt.figure(figsize=(4,3))
    plt.scatter(results['reward_gibson'], results['reward_doom'], 
#     plt.scatter(results['rank_gibson'], results['rank_doom'], 
                color=[map_to_colors[map_to_type[f]] if f in nav_sig or f in exp_sig else '#666666' for f in results.index],
                s=[80 if f in nav_sig or f in exp_sig else 20 for f in results.index])
    for i, txt in enumerate(results.index):
        if txt in ['scratch', 'blind', 'pixels_as_state']:
            y_offset = (results['rank_gibson'].max() -  results['rank_gibson'].min()) / 100 * 2
            plt.annotate(map_to_display_names[txt], (results['reward_gibson'][i] + y_offset,  results['reward_doom'][i]))
    plt.xlabel("Gibson Reward", size=12)
    plt.ylabel("Doom Reward", size=12)
    plt.title(settings['title'], size=16)
    plt.tight_layout()
    
    if plot_num == 1:
        patch2d = plt.scatter([],[], marker="o",color=map_to_colors['2d'])
        patch3d = plt.scatter([],[], marker="o", color=map_to_colors['3d'])
        patchnone = plt.scatter([],[], marker="o", color=map_to_colors['semantic'])
        patchsem = plt.scatter([],[], marker="o", color=map_to_colors['none'])
        # plt.legend(handles=[train_patch, test_patch])
        leg = plt.legend(handles=[patch2d, patch3d, patchsem, patchnone],
                   labels=['2D/Texture', '3D/Geometric', 'Semantic', 'Other'],
                   prop={'size': 8},
                   framealpha=1.0,
                   facecolor='w',
                   frameon=True)

    plt.savefig(settings['out_file'], transparent=True)
    plt.show()
```
